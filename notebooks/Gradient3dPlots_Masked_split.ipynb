```python
import numpy as np
import pandas as pd
import seaborn as sns; sns.set_style('white')
import matplotlib.pyplot as plt
import matplotlib.cm as cm
from mpl_toolkits.mplot3d import Axes3D

%matplotlib widget


df_3d = pd.read_csv('Results_Masked_split_wide.csv')

tasks = ['Semantic', 'Context']
contrasts = ['Probe_MCB', 'Probe_SCB',
             'Dots_MCB', 'Dots_SCB',
             'Decision_MCB', 'Decision_SCB',]

for task in tasks:
    for contrast in contrasts:
        fig = plt.figure()
        ax = fig.add_subplot(111, projection = '3d')
        
        ax.set_xlabel("Gradient 1")
        ax.set_ylabel("Gradient 2")
        ax.set_zlabel("Gradient 3")
        
        x_sem = df_3d['SemanticPathway_{}_{}_1'.format(task,contrast)]
        y_sem = df_3d['SemanticPathway_{}_{}_2'.format(task,contrast)]
        z_sem = df_3d['SemanticPathway_{}_{}_3'.format(task,contrast)]
        
        x_con = df_3d['ContextPathway_{}_{}_1'.format(task,contrast)]
        y_con = df_3d['ContextPathway_{}_{}_2'.format(task,contrast)]
        z_con = df_3d['ContextPathway_{}_{}_3'.format(task,contrast)]
        
        ax.scatter(x_sem, y_sem, z_sem,c='r')
        ax.scatter(x_con, y_con, z_con,c='b')
        plt.title('{}_{}'.format(task,contrast))
        plt.show()
```
