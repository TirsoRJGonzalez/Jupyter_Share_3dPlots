```python
import numpy as np
import pandas as pd
import seaborn as sns; sns.set_style('white')
import matplotlib.pyplot as plt
import matplotlib.cm as cm
from mpl_toolkits.mplot3d import Axes3D

%matplotlib widget
```

```python
df = pd.read_csv('Results.csv')
df_3d = pd.read_csv('Results_wide.csv')
c = cm.rainbow(np.linspace(0, 1, df_3d.shape[0]))
```

```python
fig = plt.figure()
ax = fig.add_subplot(111, projection = '3d')

x = df_3d['Context_Probe_SCB_1']
y = df_3d['Context_Probe_SCB_2']
z = df_3d['Context_Probe_SCB_3']

ax.set_xlabel("Gradient 1")
ax.set_ylabel("Gradient 2")
ax.set_zlabel("Gradient 3")

ax.scatter(x, y, z, color=c)

plt.title('Context Probe SCB')

plt.show()
```
